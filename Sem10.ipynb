Применить доп измерение в точечных графиках, используя аргументы hue для отношения bill_length_mm и bill_depth_mm, по sex

```python
penguins = sns.load_dataset("penguins")
penguins.head()
penguins.columns
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='sex')
```

```python
cols=['species', 'island', 'bill_length_mm', 'bill_depth_mm',
       'flipper_length_mm', 'body_mass_g', 'sex']
g=sns.PairGrid(penguins[cols])
g.map(sns.scatterplot)
```

Создать группу в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина. high - высокий(от 42), middle - средний(от 35 до 42), low - низкий(до 35).
change

```python
penguins.loc[penguins['bill_length_mm']>42, 'bill']='high'
penguins.loc[(penguins['bill_length_mm']<=42)&(penguins['bill_length_mm']>=35), 'bill']='middle'
penguins.loc[penguins['bill_length_mm']<35, 'bill']='low'
sns.histplot(data=penguins, x='bill_length_mm', hue='bill')
```

```python
sns.histplot(data=penguins, x='flipper_length_mm', hue='bill')
```
